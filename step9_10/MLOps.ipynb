```python
%%writefile "/content/drive/My Drive/MLOps_Project/gradio_app.py"
import gradio as gr
import joblib
import pandas as pd

# Load the trained model
MODEL_PATH = "Gradient_Boosting_best_model.pkl"
try:
    model = joblib.load(MODEL_PATH)
except Exception as e:
    raise RuntimeError(f"Failed to load model: {e}")

# Define the prediction function
def get_prediction(age, workclass, fnlwgt, education, education_num, marital_status,
                   occupation, relationship, race, sex, capital_gain, capital_loss,
                   hours_per_week, native_country):

    input_data = pd.DataFrame([{
        "age": age,
        "workclass": workclass,
        "fnlwgt": fnlwgt,
        "education": education,
        "education_num": education_num,
        "marital_status": marital_status,
        "occupation": occupation,
        "relationship": relationship,
        "race": race,
        "sex": sex,
        "capital_gain": capital_gain,
        "capital_loss": capital_loss,
        "hours_per_week": hours_per_week,
        "native_country": native_country
    }])

    # Get prediction
    prediction = model.predict(input_data)[0]

    # Convert prediction to readable format
    return "Income >50K" if prediction == 1 else "Income <=50K"

# Create Gradio Interface
demo = gr.Interface(
    fn=get_prediction,
    inputs=[
        gr.Number(label="Age"),
        gr.Textbox(label="Workclass"),
        gr.Number(label="Final Weight"),
        gr.Textbox(label="Education"),
        gr.Number(label="Education Number"),
        gr.Textbox(label="Marital Status"),
        gr.Textbox(label="Occupation"),
        gr.Textbox(label="Relationship"),
        gr.Textbox(label="Race"),
        gr.Textbox(label="Sex"),
        gr.Number(label="Capital Gain"),
        gr.Number(label="Capital Loss"),
        gr.Number(label="Hours per Week"),
        gr.Textbox(label="Native Country")
    ],
    outputs=gr.Textbox(label="Prediction"),
    title="Income Prediction Model",
    description="Enter your details to predict if income is greater than 50K or not."
)

# Launch Gradio App
if __name__ == "__main__":
    demo.launch(share=True)
```

```python
import pandas as pd

# Define file paths
train_data_path = "/content/drive/My Drive/MLOps_Project/train.parquet"
prod_data_path = "/content/drive/My Drive/MLOps_Project/test.parquet"

# Read the datasets
train_df = pd.read_parquet(train_data_path)
prod_df = pd.read_parquet(prod_data_path)

# Show data samples
print("Train Data Sample:")
print(train_df.head())

print("\nProduction Data Sample:")
print(prod_df.head())
```

```python
# Select only numeric columns for data drift analysis
numeric_features = train_df.select_dtypes(include=["number"]).columns

# Convert to NumPy arrays for Alibi-Detect
X_train = train_df[numeric_features].values
X_prod = prod_df[numeric_features].values

print(f"Selected numeric features: {list(numeric_features)}")
```

```python
from alibi_detect.cd import KSDrift

# Initialize the drift detector
ks_drift = KSDrift(X_train, p_val=0.05)  # 5% significance level

# Detect drift on production data
drift_result = ks_drift.predict(X_prod)

# Show the drift detection result
print(drift_result)
```

```python
import pandas as pd

# Extract drift scores
feature_names = numeric_features
p_values = drift_result["data"]["p_val"]
distances = drift_result["data"]["distance"]
drift_detected = ["Yes" if p < 0.05 else "No" for p in p_values]  # Check for drift per feature

# Create DataFrame for better readability
drift_summary = pd.DataFrame({
    "Feature": feature_names,
    "P-Value": p_values,
    "Distance": distances,
    "Drift Detected": drift_detected
})

# Show the table
# This is where your error was and was fixed
from IPython.display import display
display(drift_summary)
# instead of using a custom module, use this native function.
# It will give you a pretty-formatted output of your dataframe in notebooks
```
